Part 1

```python
import numpy as np
from matplotlib import pyplot as plt
from scipy import integrate
```

```python
t, v = np.loadtxt("droptower_vdata.txt", skiprows=5, unpack=True)
x = integrate.cumtrapz(v, x=t, initial=0)
a = np.diff(v,prepend=0)

thirties = np.full(x.shape, 30)
h = x + thirties
```

```python
f, (ax1, ax2, ax3, ax4) = plt.subplots(4,1,figsize=(10,10))

ax1.plot(t,x)
ax1.set_xlabel("Time (s)")
ax1.set_ylabel("Displacement (m)")
ax1.set_title("Displacement")
ax2.plot(t,h)
ax2.set_xlabel("Time (s)")
ax2.set_ylabel("Height (m)")
ax2.set_title("Height")
ax3.plot(t,v)
ax3.set_xlabel("Time (s)")
ax3.set_ylabel("Velocity (m/s)")
ax3.set_title("Velocity")
ax4.plot(t,a)
ax4.set_xlabel("Time (s)")
ax4.set_ylabel("Acceleration (m/s^2)")
ax4.set_title("Acceleration")

plt.tight_layout()
plt.show()
```

```python
positive_a = np.array([a_i for a_i in a if a_i > 0])
```
